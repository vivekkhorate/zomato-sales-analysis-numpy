### Zomato Sales Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 120000, 150000, 180000, 210000],  # Domino's Pizza
    [2, 100000, 130000, 160000, 190000],  # McDonald's
    [3, 90000, 120000, 150000, 170000],   # Subway
    [4, 110000, 140000, 175000, 200000],  # KFC
    [5, 80000, 100000, 130000, 160000]    # Starbucks
])

print("===Zomato Sales Analysis====")
print("\n Sales Data Shape",sales_data.shape)
print("\n Sample data for 1st 3 restaurant: ",sales_data[:,1:])
```

```python
#total sales per year
print(np.sum(sales_data, axis=0))
yeary_total=np.sum(sales_data[:,1:],axis=0)
print(yeary_total)
print(yeary_total.sum())

#Minimum sales per restaurant 
min_sales=np.min(sales_data[:,1:],axis=1)
print("\n Minimum sales per restaurant: ",min_sales)
  
#Maximum sales per restaurant 
max_sales=np.max(sales_data[:,1:],axis=0)
print("\n Maximum sales per restaurant: ",max_sales)

#Average sales per restaurant
avg_sales=np.mean(sales_data[:,1:],axis=1)      
print("\n Average sales per restaurant",avg_sales)                    
```

```python
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

plt.Figure(figsize=(7,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales accross all restaurant")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
